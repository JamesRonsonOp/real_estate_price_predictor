```python
def mass_value(df):
    # function that takes in a dataframe and returns value_counts for all columns. 
    value_list = []
    for item in df.columns:
        value_list.append(item.upper())
        value_list.append(df[item].value_counts())
    return value_list
    
```

# DUMMIES

### Create Dummies for both Df's

```python
def dummy_creator(df):
    # a function that takes a df and returns a df with categorical variables dummified. 
    strings = []
    for feature in df.columns:
        if df[feature].dtypes == str or df[feature].dtypes == object:
            strings.append(feature)
    return pd.get_dummies(df, columns = strings, drop_first = True)
```

# NULLS FINAL

### Convert or Eliminate Leftover Nulls

```python
def fill_na_mean(df):
    #Cleans float and integer NaNs by replacing with mean
    
    for item in df.columns:
        if df[item].dtypes == float or df[item].dtypes == int:
            df[item].fillna(df[item].mean(), inplace = True)
```

```python
# removing nulls from training and replacing with mean
fill_na_mean(ames_dummy)
```

```python
# confirming no nulls in training
msno.bar(ames_dummy);
```

```python
# removing nulls from test and replacing with mean. 
fill_na_mean(test_dummy)
```

```python
# confirming no nulls in test
msno.bar(test_dummy);
```
